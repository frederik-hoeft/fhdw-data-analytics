# Deskriptive Statistiken

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Excel-Datei laden
data = pd.read_excel('Spotify_data.xlsx')

#Age
# Altersverteilung plotten
plt.hist(data['Age'], bins=20, edgecolor='k')
plt.xlabel('Alter')
plt.ylabel('Anzahl')
plt.title('Altersverteilung')
plt.show()

#Gender
# Altersverteilung plotten
plt.hist(data['Gender'], bins=20, edgecolor='k')
plt.xlabel('Geschlecht')
plt.ylabel('Anzahl')
plt.title('Geschlechtsverteilung')
plt.show()

#spotify_usage_period
plt.hist(data['spotify_usage_period'], bins=20, edgecolor='k')
plt.xlabel('Usage_period')
plt.ylabel('Anzahl')
plt.title('Spotify_usage_period')
plt.show()

#spotify_listening_device
plt.xticks(rotation=45, ha='right')  # rotation in Grad und Ausrichtung 'right'
plt.hist(data['spotify_listening_device'], bins=20, edgecolor='k')
plt.xlabel('Geräte')
plt.ylabel('Anzahl')
plt.title('Geräte auf denen gehört wird')
plt.show()

#bevorzugter kontent
plt.hist(data['preferred_listening_content'], bins=20, edgecolor='k')
plt.xlabel('Bevorzugter Kontent')
plt.ylabel('Anzahl')
plt.title('Bevorzugtes User-Kontent')
plt.show()

#bevorzugtes genre
plt.xticks(rotation=45, ha='right')  # rotation in Grad und Ausrichtung 'right'
plt.hist(data['fav_pod_genre'], bins=20, edgecolor='k')
plt.xlabel('Genre')
plt.ylabel('Anzahl')
plt.title('Lieblings Genre')
plt.show()

#bevorzugtes format
plt.hist(data['preffered_pod_format'], bins=20, edgecolor='k')
plt.xlabel('Format')
plt.ylabel('Anzahl')
plt.title('Bevorzugtes Podcast-Format')
plt.show()

#host_preferenzen
plt.hist(data['pod_host_preference'], bins=20, edgecolor='k')
plt.xlabel('Preferenz')
plt.ylabel('Anzahl')
plt.title('Bevorzugtes Host-Preferenzen')
plt.show()

```

# Heatmap-übersicht mit nur den relevanten Daten

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="darkgrid")  
custom_palette = sns.color_palette("viridis")  

# Excel-Datei laden
data = pd.read_excel('Spotify_data.xlsx')
df = pd.read_excel('Spotify_data.xlsx')


# Mapping von Kategorien zu nummerischen Werten
age_mapping = {'6-12': 0,
               '12-20': 1,
               '20-35':2,
               '35-60': 3, 
               '60+': 4
              }
gender_mapping = {
    'Female': 0, 
    'Male': 1,
    'Others': 2
}
preferred_listening_content_mapping = {
    'Podcast':0,
    'Music':1
}
fav_music_genre_mapping = {
    'Melody':0,
    'Rap':1,
    'Pop':2,
    'Classical & melody':3,
    'dance':4,
    'classical':5,
    'Rock':6,
    'Old songs':7,
    'Electronic/Dance':8,
    'All':9,
    'Kpop':10,
    'trending songs random':11
}
music_time_slot_mapping = {
    'Night':0,
    'Morning':1,
    'Afternoon':2
}
music_Influencial_mood_mapping = {
    'Sadness or melancholy':0,
    'Uplifting and motivational':1,
    'Relaxation and stress relief':2,
    'Social gatherings or parties':3
}
music_expl_method_mapping = {
    'Playlists':0,
    'recommendations, Playlists':1,
    'Others':2,
    'recommendations':3,
    'Playlists, Radio':4,
    'recommendations, Others':5,
    'recommendations, Playlists, Radio':6,
    'recommendations, Radio':7,
    'recommendations, Playlists, Others':8,
    'Radio':9,
    'Radio, Others':10,
    'Playlists, Others':11,
    'Playlists, Radio, Others':12,
    'recommendations, Radio, Others':13,
    'recommendations,Others, Social media':14,
    'Others, Social media':15,
    'recommendations, Playlists, Radio, Others':16,
    'Others, Friends':17,
    'recommendations, Others, Social media':18,
    'Others, Search':19
}

pod_lis_frequency_mapping = {
    'Daily': 0,
    'Several times a week': 1,
    'Once a week': 2,
    'Never': 3,
    'Rarely': 4
}
music_lis_frequency_mapping = {

    'leisure time':0,
    'Workout session':1,
    'Study Hours':2, 
    'While Traveling':3,
    'Office hours':4,
    'Social gatherings':5,
    'Night time':6,
    'when cooking':7,
    'Random ':8,
    'Before bed ':9
}

fav_pod_genre_mapping = {
    'Comedy':0,
    'Sports':1,
    'None':2,
    'Lifestyle and Health':3,
    'Business':4,
    'Health and Fitness':5,
    'Food and cooking':6,
    'Everything':17,
    'Informative stuff':17,
    'Dance and Relevant cases':17, 
    'Technology':17,
    'General knowledge':17, 
    'Spiritual and devotional':17,
    'Finance related and current affairs':17,
    'Murder Mystery ':17,
    'Political, informative, topics that interests me':17,
    'Novels':17,
    'Self help':17,
    'Educational':17, 
    'Stories ':17
}

preffered_pod_format_mapping = {
    'Interview':0,
    'None':1,
    'Story telling':2,
    'Conversational':3,
    'Educational':4
}

pod_host_preference_mapping = {
    'Both':0,
    'None':1,
    'Well known individuals':2,
    'unknown Podcasters':3
}

preffered_pod_duration_mapping = {
    'Both':0,
    'None':1,
    'Longer':2,
    'Shorter':3
}

pod_variety_satisfaction_mapping = {
    'Ok':0,
    'Satisfied':1,
    'Very Satisfied':2,
    'Very Dissatisfied':3,
    'Dissatisfied':4
}

# Funktion zum Parsen, Mappen und Zusammenführen
# Kategorische Spalten in numerische Werte umwandeln
data['Age'] = data['Age'].map(age_mapping)
data['Gender'] = data['Gender'].map(gender_mapping)
data['preferred_listening_content'] = data['preferred_listening_content'].map(preferred_listening_content_mapping)

data['pod_lis_frequency'] = data['pod_lis_frequency'].map(pod_lis_frequency_mapping)
data['fav_pod_genre'] = data['fav_pod_genre'].map(fav_pod_genre_mapping)
data['preffered_pod_format'] = data['preffered_pod_format'].map(preffered_pod_format_mapping)

data['pod_host_preference'] = data['pod_host_preference'].map(pod_host_preference_mapping)
data['preffered_pod_duration'] = data['preffered_pod_duration'].map(preffered_pod_duration_mapping)
data['pod_variety_satisfaction'] = data['pod_variety_satisfaction'].map(pod_variety_satisfaction_mapping)

# Korrelationsmatrix berechnen
correlation_matrix = data.corr()

# Heatmap erstellen
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, cmap=custom_palette, center=0, linewidths=2)
plt.title('Korrelations-Heatmap der Daten')
plt.show()

######################################################################################################

# Count pro Kategorie ermitteln
category_counts = df['fav_pod_genre'].value_counts()

# Kategorien mit einem Count von weniger als 10 ermitteln
categories_to_combine = category_counts[category_counts < 10].index

# DataFrame kopieren
filtered_df = df.copy()

# Kategorien mit weniger als 10 Vorkommen durch "others" ersetzen
filtered_df['fav_pod_genre'] = filtered_df['fav_pod_genre'].apply(lambda x: 'others' if x in categories_to_combine else x)

sns.set_theme(style="darkgrid")  

#Diagramm mit lieblings-Genre in unterteilung des Geschlechtes
plt.figure(figsize=(10, 6))
sns.histplot(data=filtered_df, x='fav_pod_genre', hue='Gender', multiple='stack', bins=len(filtered_df['fav_pod_genre'].unique()), palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Verteilung der bevorzugten Podcast-Genres nach Geschlecht')
plt.xlabel('Bevorzugtes Podcast-Genre')
plt.ylabel('Count')

plt.show()

#############

sns.set_theme(style="darkgrid")  
custom_palette = sns.color_palette("viridis") 

#Diagramm mit lieblings-Genre in unterteilung des Alters
sns.histplot(data=filtered_df, x='fav_pod_genre', hue=df['Age'], multiple='stack', bins=len(filtered_df['fav_pod_genre'].unique()), palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Verteilung der bevorzugten Podcast-Genres nach Alter')
plt.xlabel('Bevorzugtes Podcast-Genre')
plt.ylabel('Count')

plt.show()

##############



crosstab = pd.crosstab(df['preffered_pod_duration'], df['preffered_pod_format'])

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid") 
custom_palette = sns.color_palette("viridis")  

# Heatmap der Kreuztabelle
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Kreuztabelle: bevorzugtes Format vs. bevorzugte Dauer')
plt.xlabel('Format')
plt.ylabel('Dauer')
plt.show()


################
crosstab = pd.crosstab(df['preffered_pod_duration'], df['Age'])

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid") 
custom_palette = sns.color_palette("viridis")  

# Heatmap der Kreuztabelle
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Kreuztabelle: Alter vs. bevorzugte Dauer')
plt.xlabel('Alter')
plt.ylabel('Dauer')
plt.show()

##################

# Erstelle eine Liste mit der gewünschten Reihenfolge für das Alter
alter_reihenfolge = ['6-12', '12-20', '20-35', '35-60', '60+']

# Erstelle ein kategorisches Alter-Attribut im DataFrame, um die Reihenfolge festzulegen
df['Alter_Kategorie'] = pd.Categorical(df['Age'], categories=alter_reihenfolge, ordered=True)

# Histogramm für das Alter mit der festgelegten Reihenfolge
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Alter_Kategorie', kde=True, color='blue')
plt.xticks(rotation=45, ha='right')
plt.title('Histogramm des Alters')
plt.xlabel('Alter')
plt.ylabel('Häufigkeit')
plt.show()

#############################

# Histogramm für bevorzugte Podcast-Dauer
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='preffered_pod_duration', hue='preferred_listening_content', multiple='stack', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Histogramm der bevorzugten Podcast-Dauer')
plt.xlabel('Bevorzugte Podcast-Dauer')
plt.ylabel('Häufigkeit')
plt.show()

################################

# Histogramm für Podcast-Hörhäufigkeit
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='pod_lis_frequency', hue='preferred_listening_content', multiple='stack', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Histogramm der Podcast-Hörhäufigkeit')
plt.xlabel('Podcast-Hörhäufigkeit')
plt.ylabel('Häufigkeit')
plt.show()
```
