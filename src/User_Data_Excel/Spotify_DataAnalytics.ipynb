# Kreuztabelle

```python
import pandas as pd

data = pd.read_excel('Spotify_data.xlsx')

# Beispiel: Kreuztabelle zwischen Geschlecht und Alter
crosstab = pd.crosstab(data['Gender'], data['Age'])

import matplotlib.pyplot as plt
import seaborn as sns

# Heatmap der Kreuztabelle
plt.figure(figsize=(10, 6))
sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt="d")
plt.title('Kreuztabelle: Geschlecht vs. Alter')
plt.xlabel('Alter')
plt.ylabel('Geschlecht')
plt.show()
```

# Cluster-Analyse

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

data = pd.read_excel('Spotify_data.xlsx')

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot der Elbow-Methode
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Anzahl der Cluster')
plt.ylabel('Inertia')
plt.title('Elbow-Methode')
plt.show()

num_clusters = 3  # Beispiel: Anzahl der Cluster
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(scaled_data)

cluster_labels = kmeans.labels_
data['Cluster'] = cluster_labels

sns.scatterplot(x='Feature1', y='Feature2', hue='Cluster', data=data, palette='viridis')
plt.title('Clusteranalyse')
plt.show()
```
